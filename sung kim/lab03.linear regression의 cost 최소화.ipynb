```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# optimizer를 통해 최소화
# gradient를 조절하고 싶을 때가 있음
# gvs = optimizer.compute_gradients(cost) -> 컴퓨터가 gradient 계산해줌줌# apply_gradient = optimizer.apply_gradients(gvs)
```

```python
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
```

```python
sgd = tf.keras.optimizers.SGD(lr=0.1)
tf.model.compile(loss='mse', optimizer=sgd)
```

```python
tf.model.summary()
```

```python
history = tf.model.fit(x_train, y_train, epochs=100)
```

```python
y_predict = tf.model.predict(np.array([5,4]))
print(y_predict)
```

```python
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right') #legend([라벨명], loc는 위치)
plt.show()
```
